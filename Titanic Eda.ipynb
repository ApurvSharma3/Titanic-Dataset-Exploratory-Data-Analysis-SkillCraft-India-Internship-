```python
# 🧭 TASK 2: Exploratory Data Analysis (EDA) on the Titanic Dataset

# ================================================================
# 📦 1. Import Required Libraries
# ================================================================
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

sns.set(style="whitegrid", context="talk")

# ================================================================
# 📂 2. Load the Titanic Dataset (Pre-installed in Seaborn)
# ================================================================
data = sns.load_dataset("titanic")

print("✅ Titanic dataset loaded successfully!\n")
print(f"Shape of dataset: {data.shape}\n")
print("Columns available:", list(data.columns), "\n")

# ================================================================
# 🔍 3. Initial Exploration
# ================================================================
print("Dataset Overview:\n")
display(data.head())

print("\nData Types:\n", data.dtypes)
print("\nMissing Values:\n", data.isnull().sum())

# ================================================================
# 🧹 4. Data Cleaning
# ================================================================
data['age'].fillna(data['age'].median(), inplace=True)
data['embarked'].fillna(data['embarked'].mode()[0], inplace=True)
data['deck'].fillna('Unknown', inplace=True)
data.dropna(subset=['embark_town'], inplace=True)

print("\nMissing values handled successfully.\n")

# ================================================================
# 📊 5. Descriptive Statistics
# ================================================================
print("Summary Statistics:\n")
display(data.describe(include='all').T)

# ================================================================
# 🌍 6. Univariate Analysis
# ================================================================
# Survival Distribution
plt.figure(figsize=(6,6))
colors = ["#E74C3C", "#27AE60"]
plt.pie(data['survived'].value_counts(), 
        labels=['Not Survived', 'Survived'], 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=colors)
plt.title("Overall Survival Distribution", fontsize=18, fontweight="bold")
plt.show()

# ================================================================
# 👩‍🦱 7. Survival by Gender
# ================================================================
plt.figure(figsize=(8,6))
sns.countplot(x="sex", hue="survived", data=data, palette=colors)
plt.title("Survival Count by Gender", fontsize=18, fontweight="bold")
plt.xlabel("Gender")
plt.ylabel("Passenger Count")
plt.legend(labels=["Not Survived", "Survived"])
plt.show()

# ================================================================
# 💺 8. Survival by Passenger Class
# ================================================================
plt.figure(figsize=(8,6))
sns.countplot(x="pclass", hue="survived", data=data, palette=colors)
plt.title("Survival Count by Passenger Class", fontsize=18, fontweight="bold")
plt.xlabel("Passenger Class (1 = Upper, 3 = Lower)")
plt.ylabel("Passenger Count")
plt.legend(labels=["Not Survived", "Survived"])
plt.show()

# ================================================================
# 🎂 9. Age Distribution
# ================================================================
plt.figure(figsize=(10,6))
sns.histplot(data, x="age", bins=30, kde=True, color="#3498DB")
plt.title("Age Distribution of Passengers", fontsize=18, fontweight="bold")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# ================================================================
# 🧒 10. Age vs Survival
# ================================================================
plt.figure(figsize=(10,6))
sns.kdeplot(data=data[data['survived']==1]['age'], shade=True, color="#27AE60", label="Survived")
sns.kdeplot(data=data[data['survived']==0]['age'], shade=True, color="#E74C3C", label="Not Survived")
plt.title("Age Distribution by Survival", fontsize=18, fontweight="bold")
plt.xlabel("Age")
plt.ylabel("Density")
plt.legend()
plt.show()

# ================================================================
# 💰 11. Fare vs Class
# ================================================================
plt.figure(figsize=(10,6))
sns.boxplot(x="pclass", y="fare", data=data, palette="coolwarm")
plt.title("Fare Distribution by Passenger Class", fontsize=18, fontweight="bold")
plt.xlabel("Passenger Class")
plt.ylabel("Fare ($)")
plt.show()

# ================================================================
# 🧭 12. Correlation Heatmap
# ================================================================
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Heatmap of Numeric Variables", fontsize=18, fontweight="bold")
plt.show()

# ================================================================
# ⚡ 13. Interactive Scatter: Survival by Age & Fare
# ================================================================
fig = px.scatter(
    data_frame=data,
    x="age",
    y="fare",
    color="survived",
    size="fare",
    hover_data=["sex", "class", "embarked"],
    title="Survival by Age, Fare & Class (Interactive)",
    color_continuous_scale=["red", "green"]
)
fig.show()
```

```python
# 🛳️ TASK 2: Titanic Dataset - Exploratory Data Analysis (Professional & Visual Edition)
# ------------------------------------------------------------

# 📦 Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Visualization Style
sns.set_theme(style="whitegrid", context="talk", palette="deep")
plt.rcParams["figure.facecolor"] = "#f9f9f9"
plt.rcParams["axes.facecolor"] = "#f9f9f9"

# ------------------------------------------------------------
# 📂 Load Titanic Dataset
data = sns.load_dataset("titanic")
print("✅ Titanic dataset loaded successfully!")
display(data.head())

# ------------------------------------------------------------
# 🧹 Basic Cleaning (Fixed deck issue)
data['age'].fillna(data['age'].median(), inplace=True)
data['embarked'].fillna(data['embarked'].mode()[0], inplace=True)
data['deck'] = data['deck'].astype(str).fillna('Unknown')  # ✅ Fixed line
data.dropna(subset=['embark_town'], inplace=True)

# ------------------------------------------------------------
# 🎨 Define Custom Palette
palette_main = ["#E74C3C", "#27AE60", "#3498DB", "#F1C40F", "#8E44AD", "#2ECC71"]

# ------------------------------------------------------------
# 📊 1. Overall Survival Distribution
fig, ax = plt.subplots(figsize=(6,6))
colors = ["#E74C3C", "#2ECC71"]
wedges, texts, autotexts = ax.pie(
    data['survived'].value_counts(),
    labels=["Did Not Survive", "Survived"],
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
    wedgeprops={"edgecolor": "white", "linewidth": 2}
)
plt.setp(autotexts, size=15, weight="bold", color="white")
ax.set_title("🧍 Overall Survival Distribution", fontsize=18, weight="bold")
plt.show()

# ------------------------------------------------------------
# 👩‍🦱 2. Gender vs Survival
plt.figure(figsize=(8,6))
sns.countplot(data=data, x="sex", hue="survived", palette=colors, edgecolor="black")
plt.title("👩‍🦰 Survival Count by Gender", fontsize=18, weight="bold")
plt.xlabel("Gender", fontsize=14)
plt.ylabel("Passenger Count", fontsize=14)
plt.legend(labels=["Did Not Survive", "Survived"], title="Status")
sns.despine()
plt.show()

# ------------------------------------------------------------
# 💺 3. Passenger Class vs Survival
plt.figure(figsize=(8,6))
sns.countplot(data=data, x="pclass", hue="survived", palette=colors, edgecolor="black")
plt.title("💺 Survival Count by Passenger Class", fontsize=18, weight="bold")
plt.xlabel("Passenger Class (1 = Upper, 3 = Lower)", fontsize=14)
plt.ylabel("Passenger Count", fontsize=14)
plt.legend(labels=["Did Not Survive", "Survived"], title="Status")
sns.despine()
plt.show()

# ------------------------------------------------------------
# 🎂 4. Age Distribution
plt.figure(figsize=(10,6))
sns.histplot(data=data, x="age", bins=30, kde=True, color="#3498DB")
plt.title("🎂 Age Distribution of Passengers", fontsize=18, weight="bold")
plt.xlabel("Age", fontsize=14)
plt.ylabel("Count", fontsize=14)
sns.despine()
plt.show()

# ------------------------------------------------------------
# 🧒 5. Age vs Survival Density
plt.figure(figsize=(10,6))
sns.kdeplot(data=data[data['survived']==1]['age'], shade=True, color="#27AE60", label="Survived", alpha=0.7)
sns.kdeplot(data=data[data['survived']==0]['age'], shade=True, color="#E74C3C", label="Did Not Survive", alpha=0.7)
plt.title("🧒 Age Distribution by Survival", fontsize=18, weight="bold")
plt.xlabel("Age")
plt.ylabel("Density")
plt.legend()
sns.despine()
plt.show()

# ------------------------------------------------------------
# 💰 6. Fare Distribution by Class
plt.figure(figsize=(10,6))
sns.boxplot(data=data, x="pclass", y="fare", palette="coolwarm", linewidth=2)
plt.title("💰 Fare Distribution Across Passenger Classes", fontsize=18, weight="bold")
plt.xlabel("Passenger Class")
plt.ylabel("Fare ($)")
sns.despine()
plt.show()

# ------------------------------------------------------------
# 🌍 7. Embarkation Town vs Survival
plt.figure(figsize=(8,6))
sns.countplot(data=data, x="embark_town", hue="survived", palette=colors, edgecolor="black")
plt.title("🌍 Survival Count by Embarkation Town", fontsize=18, weight="bold")
plt.xlabel("Embarkation Town")
plt.ylabel("Passenger Count")
plt.legend(labels=["Did Not Survive", "Survived"], title="Status")
sns.despine()
plt.show()

# ------------------------------------------------------------
# ⚡ 8. Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="crest", fmt=".2f", linewidths=1, cbar_kws={"shrink":0.8})
plt.title("⚡ Correlation Heatmap (Numeric Variables)", fontsize=18, weight="bold")
sns.despine()
plt.show()

# ------------------------------------------------------------
# 🧭 9. Interactive Plot - Age vs Fare vs Survival
fig = px.scatter(
    data_frame=data,
    x="age",
    y="fare",
    color="survived",
    size="fare",
    hover_data=["sex", "class", "embark_town"],
    color_continuous_scale=["#E74C3C", "#2ECC71"],
    title="🧭 Age vs Fare vs Survival (Interactive)",
    labels={"survived": "Survival Status"},
    template="plotly_white"
)
fig.update_traces(marker=dict(line=dict(width=1, color="DarkSlateGrey")))
fig.show()
```
